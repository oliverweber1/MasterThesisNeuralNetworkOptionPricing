```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class StochasticProcess:
    
    def __init__(self, T, steps):
        self.time = np.linspace(0, T, steps)
        self.T = T
```

```python
class CompoundPoissonProcess(StochasticProcess):
    
    def __init__(self, T=10, steps=10000, lam=1, jumpSizeRV=sp.norm, drift=0):
        super().__init__(T, steps)
        self.lam = lam
        self.jumpSizeRV = jumpSizeRV
        self.jumpNumberRV = sp.poisson(mu = lam * T)
        self.drift = drift
        
    def generatePath(self):
        numberOfJumps = self.jumpNumberRV.rvs()
        jumpTimes = np.append(sorted(sp.uniform.rvs(scale = self.T, size = numberOfJumps)), self.T+1)
        jumpValues = np.append(0, np.cumsum(self.jumpSizeRV.rvs(size = numberOfJumps)))
        path = []
        jumpIndex = 0
        for i in range(len(self.time)):
            if self.time[i] < jumpTimes[jumpIndex]:
                path.append(self.drift * self.time[i] + jumpValues[jumpIndex])
            else:
                jumpIndex += 1
                path.append(self.drift * self.time[i] + jumpValues[jumpIndex])
        return path
    
    def plotPath(self):
        plt.figure(figsize=(10,5))
        plt.plot(self.time, self.generatePath())
        plt.xlabel('$t$')
        plt.ylabel('$N_t$')
        plt.title('Sample path of Compound Poisson Process with intensity $\lambda=${} and drift $b=${}'.format(self.lam,self.drift))
        
```

```python
process = CompoundPoissonProcess(10, 20000, 10, sp.norm(scale=5), 1)
```

```python
process.plotPath()
```
