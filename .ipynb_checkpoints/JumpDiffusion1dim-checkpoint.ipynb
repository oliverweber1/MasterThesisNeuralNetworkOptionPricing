```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class StochasticProcess:
    
    def __init__(self, T, steps, x0):
        self.time = np.linspace(0, T, steps)
        self.T = T
        self.x0 = x0
        
    def generatePath(self):
        raise NotImplementedError("Must override generatePath")
        
    def plotPath(self, numberOfPaths):
        plt.figure(figsize=(10,5))
        for _ in range(numberOfPaths):
            plt.plot(self.time, self.generatePath(), linestyle='None', marker='o', markersize='0.3')
        plt.xlabel('$t$')
        plt.ylabel('$N_t$')
        
    def getFinalValue(self):
        raise NotImplementedError("Must override getFinalValue")
```

```python
class CompoundPoissonProcess(StochasticProcess):
    
    def __init__(self, T=10, steps=10000, lam=1, jumpSizeRV=sp.norm, drift=0, x0=0):
        super().__init__(T, steps, x0)
        self.lam = lam
        self.jumpSizeRV = jumpSizeRV
        self.jumpNumberRV = sp.poisson(mu = lam * T)
        self.drift = drift
        
    def generatePath(self):
        numberOfJumps = self.jumpNumberRV.rvs()
        jumpTimes = np.append(sorted(sp.uniform.rvs(scale = self.T, size = numberOfJumps)), self.T+1)
        jumpValues = np.cumsum(np.append(self.x0, self.jumpSizeRV.rvs(size = numberOfJumps)))
        path = []
        jumpIndex = 0
        for i in range(len(self.time)):
            if self.time[i] < jumpTimes[jumpIndex]:
                path.append(self.drift * self.time[i] + jumpValues[jumpIndex])
            else:
                jumpIndex += 1
                path.append(self.drift * self.time[i] + jumpValues[jumpIndex])
        return np.array(path)
    
    def plotPath(self, numberOfPaths=1):
        super().plotPath(numberOfPaths)
        plt.title('Compound Poisson Process with intensity $\lambda=${} and drift $b=${}'.format(self.lam,self.drift))
        
    def getFinalValue(self):
        numberOfJumps = self.jumpNumberRV.rvs()
        return self.x0 + self.T * self.drift + sum(self.jumpSizeRV.rvs(size = numberOfJumps))
        
```

```python
class BrownianMotion(StochasticProcess):
    
    def __init__(self, T=10, steps=10000, mu=0, sig=1, x0=0):
        super().__init__(T, steps, x0)
        self.mu = mu
        self.sig = sig
        
    def generatePath(self):
        dt = self.T / (len(self.time) - 1)
        increments = sp.norm.rvs(scale = self.sig * np.sqrt(dt), size = len(self.time)-1)
        return self.time * self.mu + np.cumsum(np.append(self.x0, increments))
    
    def plotPath(self, numberOfPaths=1):
        super().plotPath(numberOfPaths)
        plt.title('Brownian Motion with drift $\mu=${} and diffusion $\sigma=${}'.format(self.mu,self.sig))
        
    def getFinalValue(self):
        return self.x0 + self.T * self.mu + sp.norm.rvs(scale = self.sig * np.sqrt(self.T))
        
```

```python
class JumpDiffusionProcess(StochasticProcess):
    
    def __init__(self, T=10, steps=10000, drift=0, sig=1, lam=1, jumpSizeRV=sp.norm, x0=0):
        super().__init__(T, steps, x0)
        self.brownianPart = BrownianMotion(T=T, steps=steps, mu=drift, sig=sig, x0=x0)
        self.poissonPart = CompoundPoissonProcess(T=T, steps=steps, lam=lam, jumpSizeRV=jumpSizeRV, x0=0)
        
    def generatePath(self):
        return self.brownianPart.generatePath() + self.poissonPart.generatePath()
    
    def plotPath(self, numberOfPaths=1):
        super().plotPath(numberOfPaths)
        plt.title('Jump Diffusion with drift $\mu=${}, diffusion $\sigma=${} and jump rate $\lambda=${}'.format(self.brownianPart.mu,self.brownianPart.sig,self.poissonPart.lam))
        
    def getFinalValue(self):
        return self.brownianPart.getFinalValue() + self.poissonPart.getFinalValue()
```

```python
class ExponentialJumpDiffusion(JumpDiffusionProcess):
    
    def __init__(self, T=10, steps=10000, drift=0, sig=1, lam=1, jumpSizeRV=sp.norm, s0=0):
        super().__init__(T=T, steps=steps, drift=drift, sig=sig, lam=lam, jumpSizeRV=jumpSizeRV, x0=0)
        self.s0 = s0
        
    def generatePath(self):
        return self.s0 * np.exp(super().generatePath())
    
    def plotPath(self, numberOfPaths=1):
        StochasticProcess.plotPath(self, numberOfPaths)
        plt.title('Exponential Jump Diffusion with drift $\mu=${}, diffusion $\sigma=${} and jump rate $\lambda=${}'.format(self.brownianPart.mu,self.brownianPart.sig,self.poissonPart.lam))

    def getFinalValue(self):
        return self.s0 * np.exp(super().getFinalValue())
```
